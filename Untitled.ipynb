My purpose in this project is to analyze the SAT data for the city of New York and see if there are any correlations between sub-optimal SAT scores and demographic realities such as race, region (primarily burroughs), etc. New York City is said to be one of the most ethnically diverse cities due to it being the historical hub for migraints, so it would produce an adequately representative sample by accounting for a hugh number of ethnicities, races, etc.

We will be using a number of datasets to attain the desired results. They are linked below and are open to be used by everybody:
- #### [SAT scores by School](https://data.cityofnewyork.us/Education/SAT-Results/f9bf-2cp4) 
Average SAT scores outlined for each highschool present in NYC for the year of 2022.
- #### [School attendance](https://data.cityofnewyork.us/Education/School-Attendance-and-Enrollment-Statistics-by-Dis/7z8d-msnt)
Attendances based on all the highschools present in a district for the year 2010-2011.
- #### [Class size](https://data.cityofnewyork.us/Education/2010-2011-Class-Size-School-level-detail/urz7-pzb3)
Average class sizes for each school by grade and program type.
- #### [AP test results](https://data.cityofnewyork.us/Education/AP-College-Board-2010-School-Level-Results/itfs-ms3e)
2010 AP school level results.
- #### [Graduation outcomes](https://data.cityofnewyork.us/Education/Graduation-Outcomes-Classes-Of-2005-2010-School-Le/vh2h-md7a)
Graduation outcomes for the classes of 2005-2010
- #### [Demographics](https://data.cityofnewyork.us/Education/School-Demographics-and-Accountability-Snapshot-20/ihfw-zy9j)
Annual accounts for 2006-2012 of NYC public school student populations by grades, ethnicities, gender, etc.
- #### [School Survey](https://data.cityofnewyork.us/Education/NYC-School-Survey-2011/mnz3-dyi8)
2011 surveys by the teachers, parents and students in grade 6-12 regarding the strengths and weaknesses of their respective schools.


```python
import pandas as pd
import matplotlib.pyplot as plt
import re
%matplotlib inline

data_files = [
    "ap_2010.csv",
    "class_size.csv",
    "demographics.csv",
    "graduation.csv",
    "hs_directory.csv",
    "sat_results.csv"
]
data = {}
for names in data_files:
    file_name = names.split('.')[0]
    dataframe_name = file_name + '_df'
    dataframe_name = pd.read_csv(names)
    data[file_name] = dataframe_name 
    
all_survey = pd.read_csv('survey_all.txt', encoding='windows-1252', delimiter='\t')
d75_survey = pd.read_csv('survey_d75.txt', encoding='windows-1252', delimiter='\t')
survey = pd.concat([all_survey, d75_survey], axis=0)

#Filtering out all the unneccesary columns from survey.
col_to_keep = ["DBN", "rr_s", "rr_t", "rr_p", "N_s", "N_t", "N_p", "saf_p_11", "com_p_11", "eng_p_11", "aca_p_11", "saf_t_11", "com_t_11", "eng_t_11", "aca_t_11", "saf_s_11", "com_s_11", "eng_s_11", "aca_s_11", "saf_tot_11", "com_tot_11", "eng_tot_11", "aca_tot_11"]

survey['DBN']=survey['dbn']
survey.drop('dbn', inplace=True, axis=1)
survey = survey[col_to_keep]
data['survey'] = survey
```

```python
data['sat_results'].rename(columns={'Num of SAT Test Takers': 'test_takers', 'SAT Critical Reading Avg. Score': 'reading_scores', 'SAT Math Avg. Score': 'math_scores', 'SAT Writing Avg. Score': 'writing_scores'}, inplace=True)
```

The main thing we are interested in is the SAT data. The columns in SAT dataset are explained below:
- DBN: The district burough number.
- school_name
- test_takers: Total students who took the test.
- reading_scores: SAT average critical reading scores per school.
- math_scores: SAT average critical math scores per school.
- writing_scores: SAT average critical writing scores per school.

The dataset has about 478 rows, representing 478 highschools present in NYC. The 'DBN' column can serve to be a unique identifier for each of the school since all the datasets have the column of DBN in them, except for class_size dataset.

The class_size dataset does have two columns in it which can be combined and formatted to create a DBN column for them. Those two columns are 'CSD' and 'SCHOOL CODE'. If we can get the 'CSD' columns to have a padding of 0 for some of the single digit values, we can combine 'CSD' and 'SCHOOL CODE', which is what we will be doing in the cell below.



```python
CSD = data['class_size']['CSD']

def add_pad(number):
    return '{:02d}'.format(number)


CSD = CSD.apply(add_pad)
data['class_size']['CSD'] = CSD
data['class_size']['DBN'] = data['class_size']['CSD'] + data['class_size']['SCHOOL CODE']
```

### Adding necessary columns.
Since the dataset 'sat_results' doesn't have a total SAT score column, we would add one below. We would also extract the latitudes and logtitudes for each school present in the 'hs_directory' dataset and assign it to a column called 'lat' and 'lon', respectively, in the same dataset.

```python
for cols in data['sat_results'].columns[3:]:
    data['sat_results'][cols] = pd.to_numeric(data['sat_results'][cols], errors='coerce') 

    
data['sat_results']['sat_score'] = data['sat_results'].iloc[:,3] + data['sat_results'].iloc[:,4] + data['sat_results'].iloc[:,5]

def get_lat(address):
    latitude = re.findall("\(.+\)", address)[0].split(',')[0]
    latitude = latitude.replace('(', '')
    return latitude


def get_lon(address):
    longitude = re.findall("\(.+\)", address)[0].split(',')[1]
    longitude = longitude.replace(')', '')
    return longitude

data['hs_directory']['lat'] = data['hs_directory']['Location 1'].apply(get_lat)
data['hs_directory']['lon'] = data['hs_directory']['Location 1'].apply(get_lon)

#Convert lat and lon columns to numeric datatype.
data['hs_directory']['lat'] = pd.to_numeric(data['hs_directory']['lat'], errors='coerce')
data['hs_directory']['lon'] = pd.to_numeric(data['hs_directory']['lon'], errors='coerce')

data['hs_directory']['DBN']=data['hs_directory']['dbn']
data['hs_directory'].drop('dbn', inplace=True, axis=1)
```

### Cleaning 'class_size' dataset.
The 'class_size' dataset has multiple values for the same DBN number and merging the datasets would prove to be problematic because of this. So we would be reducing the number of rows such that there are only unique DBN values. The way we will achieve this is by indexing the rows for the value '09-12' in the 'GRADE ' column and then for the value 'GEN ED' in the 'PROGRAM TYPE' column, that way we are only left with the most relevant data; the general education high schoolers.

After we have reduced the number of rows through the methods above, we are still left with repeated DBN numbers due to the fact that there are multiple rows based on different subjects and courses that a student can take for the same school. We would work around this issue by aggregating (averaging) the different 'AVERAGE CLASS SIZE' values based on the DBN numbers.

```python
class_size = data['class_size']
class_size = class_size[class_size['GRADE '] == '09-12']
class_size = class_size[class_size['PROGRAM TYPE'] == 'GEN ED']

class_size = class_size.groupby('DBN', as_index=False)['AVERAGE CLASS SIZE'].mean()
data['class_size'] = class_size
```

### Cleaning 'demographics' dataset.

This dataset also has repeating values for the DBN column because it has the demographics information for multiple school years. We will concern ourselves with only the most recent year i.e. 2011-2012. Below, we would be indexing the rows based on the values that have '20112012' as values in the 'schoolyear' column.

```python
data['demographics'] = data['demographics'][data['demographics']['schoolyear']== 20112012]
```

### Cleaning 'graduation' dataset.

The columns stopping this dataset from having unique DBN values are the 'Cohort' columns which underline the year the cohort graduated in ass well as the 'Demographic' columns where they are distinguished between different ethnicities, genders, etc. The 'Demographic' column also has 'Total Cohort' values which is what we will be going by as that includes the total cohort, rather than seperating them by some standard. 

```python
data['graduation'] = data['graduation'][data['graduation']['Cohort'] == '2006']
data['graduation'] = data['graduation'][data['graduation']['Demographic'] == 'Total Cohort']
```

### Converting columns in 'ap_2010' to numeric datatype(s).

The dataset 'ap_2010' has a few quantitative columns with 'object' datatypes. We would like to convert them to numeric datatypes so they can later be used for statistical analysis.

```python
cols = ['AP Test Takers ', 'Total Exams Taken', 'Number of Exams with scores 3 4 or 5']

for col in cols:
    data['ap_2010'][col] = pd.to_numeric(data['ap_2010'][col], errors='coerce')
```

```python
print(data['graduation'].shape)
print(data['sat_results'].shape)
print(data['class_size'].shape)
print(data['demographics'].shape)
print(data['survey'].shape)
print(data['hs_directory'].shape)
```

### Starting the merging process.
This is where we begin merging our datasets. We still have quite a few discrepancies between different datasets. Some datasets are lacking DBN numbers that are present in the 'sat_result' datasets, while some datasets have DBN rows absent in 'sat_result'. Therefore, we will take the combination of 'sat_result' with each of the other datasets on a case by case basis. For some datasets, we will use a left merge where the discrepancy between the DBN columns are huge, whereas we will use an inner join where it is not.

We will start by joining 'ap_2010' and 'graduation' to a copy of 'sat_result' with a left join, given the fact that these 2 datasets have significantly fewer rows as compared to 'sat_result'; 'ap_2010' has about half the amount and 'graduation' has about 80% the amount of 'sat_results'.

We will then move forward to doing the inner joins with the rest of the datasets as the sizes of these datasets are more similar or significantly bigger than 'sat_results' dataset.

Thereafter, we will be imputating null values primarily with the mean of the columns the null values are in, and then replacing non-numeric empty values with a 0.

Finally, we would be adding a school district column to our dataset, as it would be useful to determine the distribution of SAT scores based on the district of the city.

```python
combined = data["sat_results"]
combined = combined.merge(data['ap_2010'], on='DBN', how='left')
combined = combined.merge(data['graduation'], on='DBN', how='left')

combined = combined.merge(data['class_size'], on='DBN', how='inner')
combined = combined.merge(data['demographics'], on='DBN', how='inner')
combined = combined.merge(data['survey'], on='DBN', how='inner')
combined = combined.merge(data['hs_directory'], on='DBN', how='inner')

combined.fillna(combined.mean(), inplace=True)
combined.fillna(0, inplace=True)

def first_two(string):
    return string[0:2]


combined['school_dist'] = combined['DBN'].apply(first_two)
```

As for the first 6 columns, they correspond to response rate (rr) of the demographics and number (N) from the demographics who participated in the survey.

Below we will see whether there's any correlation between the demographics's perception of the above mentioned parameters with the schools and their SAT scores.

### Determining Correlations Between Races and SAT Score.
One of the most contested controversies around the design of SAT is that it is generally biased towards the white population. Even before we divulge in a statistical analysis of the data, a few things we can infer just through rationalizing:
- A major part of the SAT score is determined based on one's fluency and comprehension of the English language, so it is bound to be a huge obstacle for migraint populace with little to no background in English. That's surely a hurdle, but not one that should be removed for the 'convenience' of people. The medium language for education in America is English, so a lack of command over English is going to be a much bigger disadvantage and would lead to a larger waste of resources than a significant portion of demographic not achieving better SAT scores.
- A third of the SAT score is based of off mathematical abilities and competence. An argument can be made since mathematical reasoning requires the ability to comprehend, construct and deconstruct abstractions and ideas, and people living in sub-optimal conditions haven't had the privilege to develop those skills due to the immediacy of more pressing needs, SAT tests are unjust to that portion of the populace. Here again, we refer to the fact that the capability to reason through things deductively and inductively above a foundational threshold is the bare minimum required if a person wishes to pursue a STEM field. Reason being, the bare minimum is what the rest of the more complex and intricate sturctures of knowledge and understanding will be built upon and without it, the person will be lost and the institutions would have wasted their resources.

Regardless, we are yet to the nature of the correlation, so let's us delve into it!
